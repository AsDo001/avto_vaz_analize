```python
# @title **Себестоимость реализации**

df['Себестоимость_реализации_млн_до_замены'] = df['Себестоимость_реализации_млн']


median_value_0 = df['Себестоимость_реализации_млн'].median()
df['Себестоимость_реализации_млн'].fillna(median_value_0, inplace=True)


fig, ax = plt.subplots(1, 2, figsize=(12, 6))


ax[0].bar(df.index, df['Себестоимость_реализации_млн_до_замены'], color='red')
ax[0].set_title('График себестоимости реализации до замены значений')
ax[0].set_xlabel('Года')
ax[0].set_ylabel('Себестоимость (млн)')
ax[0].grid(axis='y', alpha=0.75)


ax[1].bar(df.index, df['Себестоимость_реализации_млн'], color='green')
ax[1].set_title('График себестоимости реализации после замены значений')
ax[1].set_xlabel('Года')
ax[1].set_ylabel('Себестоимость (млн)')
ax[1].grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
